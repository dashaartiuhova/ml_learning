1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = np.linspace(0, 10,100)
f = 10 * x -5

df = np.full_like(x, 10)


plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')
```

Похідна є сталою тому й кількість годин рівномірно збільшують оцінку.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_1 = sp.sin(x)
derivative_f_1 = sp.diff(f_1, x)

f_num = sp.lambdify(x, f_1, 'numpy')
df_num = sp.lambdify(x, derivative_f_1, 'numpy')
x_vals = np.linspace(0, 10, 100)

y_vals = f_num(x_vals)
dy_vals = df_num(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x) = cos(x)", color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
f_2 = sp.log(x)
derivative_f_2 = sp.diff(f_2, x)

f_num = sp.lambdify(x, f_2, 'numpy')
df_num = sp.lambdify(x, derivative_f_2, 'numpy')
x_vals = np.linspace(0, 10, 100)

y_vals = f_num(x_vals)
dy_vals = df_num(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x) = 1/x", color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
f_3 = x**2 - 5*x + 1
derivative_f_3 = sp.diff(f_3, x)

f_num = sp.lambdify(x, f_3, 'numpy')
df_num = sp.lambdify(x, derivative_f_3, 'numpy')

x_vals = np.linspace(0, 100, 100)
y_vals = f_num(x_vals)
dy_vals = df_num(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x) = 2x - 5", color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
