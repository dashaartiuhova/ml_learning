```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
print(B.T)
```

```python
B_T = B.T
result_matrix = np.dot(A, B_T)
print('Result of A dot B:\n', result_matrix)
```

```python
result_matrix = np.dot(B, C)
print('Result of B dot C:\n', result_matrix)
```
